```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
floats = pd.read_csv("Screenshots/BasketSizes_Floats/data.txt",delimiter=", ")
floats.head()
```

```python
fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2, figsize=(14,8))

floats.plot(x="basketsize", y="TTreeCompressionFactor", ax=ax1, color='dodgerblue')
floats.plot(x="basketsize", y="VpxBranchCompression", ax=ax2, color='dodgerblue')
sns.barplot(floats, x="basketsize", y="DiskMean", ax=ax3, color='dodgerblue')
sns.barplot(floats, x="basketsize", y="RMS", ax=ax4, color='dodgerblue')
floats.plot(x="basketsize", y="RealTime", ax=ax5, color='dodgerblue')
floats.plot(x="basketsize", y="CPUTime", ax=ax6, color='dodgerblue')
plt.tight_layout()
plt.savefig("baskets_floats.png");
```
